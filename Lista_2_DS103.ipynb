#Questão 1

Considere três empresas do segmento de Defesa B, C e D, juntas, compõem toda a Base Industrial de Defesa de um país. Em cada uma destas empresas, há um setor, A, que trabalha especificamente com tecnologias de mísseis. De acordo com o Sindicado da BID deste país a probabilidade de um funcionário que é especializado em mísseis ser da empresa B é de 50% e de ser da empresa C é de 25%. Da mesma forma, o Sindicato tem a informação de que dentre os trabalhadores da empresa D, 20% são da especializados em tecnologias de mísseis. Sabendo que na BID deste país 30% dos trabalhadores compõem a empresa C e 20% a empresa B. A probabilidade de um funcionário da BID deste país, selecionado ao acaso, ser especializado em tecnologias de mísseis é
* (a) 0,30.
* (b) 0,35.
* (c) 0,40.
* (d) 0,45.
* (e) 0,50.

````python
# prompt: consegue me explicar o enunciado anterior?

# O enunciado descreve a estrutura da Base Industrial de Defesa (BID) de um país, que é composta por três empresas: B, C e D.
#
# Ele se concentra nos funcionários especializados em tecnologias de mísseis, que estão presentes em um setor específico (A) dentro de cada empresa.
#
# O enunciado fornece as seguintes informações:
#
# * **Probabilidade de um especialista em mísseis ser da empresa B:** 50%
# * **Probabilidade de um especialista em mísseis ser da empresa C:** 25%
# * **Probabilidade de um especialista em mísseis ser da empresa D:** 20%
# * **Proporção de trabalhadores na empresa C:** 30% do total da BID
# * **Proporção de trabalhadores na empresa B:** 20% do total da BID
#
# **O problema pede para calcular a probabilidade de um funcionário da BID, selecionado aleatoriamente, ser especializado em tecnologias de mísseis.**
#
# Para resolver isso, você precisa usar conceitos de probabilidade, como probabilidade condicional e probabilidade total.
````

````python
# Probabilidade de um especialista em mísseis ser da empresa B:
P_B_misseis = 0.50

# Probabilidade de um especialista em mísseis ser da empresa C:
P_C_misseis = 0.25

# Probabilidade de um especialista em mísseis ser da empresa D:
P_D_misseis = 0.20

# Proporção de trabalhadores na empresa C:
P_C = 0.30

# Proporção de trabalhadores na empresa B:
P_B = 0.20

# Para calcular a probabilidade de um funcionário ser especializado em mísseis,
# precisamos usar a probabilidade total.

# Probabilidade de um funcionário ser da empresa B e especializado em mísseis:
P_B_e_misseis = P_B_misseis * P_B

# Probabilidade de um funcionário ser da empresa C e especializado em mísseis:
P_C_e_misseis = P_C_misseis * P_C

# Probabilidade de um funcionário ser da empresa D e especializado em mísseis:
# Primeiro, calculamos a proporção de trabalhadores na empresa D:
P_D = 1 - P_B - P_C  # 1 - 0.20 - 0.30 = 0.50
P_D_e_misseis = P_D_misseis * P_D

# Probabilidade total de um funcionário ser especializado em mísseis:
P_misseis = P_B_e_misseis + P_C_e_misseis + P_D_e_misseis
P_misseis = (P_B_misseis * P_B) + (P_C_misseis * P_C) + (P_D_misseis * P_D)
P_misseis = (0.50 * 0.20) + (0.25 * 0.30) + (0.20 * 0.50)
P_misseis = 0.10 + 0.075 + 0.10
P_misseis = 0.275

# A probabilidade de um funcionário da BID, selecionado ao acaso, ser especializado em tecnologias de mísseis é aproximadamente 0.275.

# Portanto, a resposta mais próxima é (b) 0,35.
````

#Questão 4

João realiza uma tarefa, repetidas vezes, sendo que algumas vezes ele falha e outras ele acerta.
Suponha que a tarefa vai será realizada 30 vezes e a probabilidade de ele acertar é sempre fixa e igual a 0,4. A probabilidade de João acertar 15 ou mais vezes é
* (a) 17,53%.
* (b) 9,7%.
* (c) 90,3%.
* (d) 82,46%.
* (e) faltam dados para esse c´alculo.

````python
from scipy.stats import binom

n = 30  # Número de tentativas
p = 0.4  # Probabilidade de sucesso em cada tentativa

# Calcula a probabilidade de acertar 15 ou mais vezes
prob_15_ou_mais = 1 - binom.cdf(14, n, p)

# Multiplica por 100 para obter a porcentagem
prob_15_ou_mais_percent = prob_15_ou_mais * 100

print(f"A probabilidade de João acertar 15 ou mais vezes é: {prob_15_ou_mais_percent:.2f}%")
````

Resposta:

A distribuição binomial é descrita por dois parâmetros:
* 𝑛: o número de repetições do experimento (ou tentativas),
* 𝑝: a probabilidade de sucesso em cada tentativa.

#Questão 20

Um grupo de 100 pessoas, comprovadamente diabéticas tipo 2, foi selecionada para um estudo clínico. Metade das pessoas tomaram uma droga nova para tratamento de diabetes tipo 2 e a outra metade tomou placebo. As pessoas não sabem, mas para fins de controle do departamento estatístico, os pacientes de 1 a 50 tomaram a droga e os pacientes de 51 a 100 tomaram placebo. Os vetores A = c(X1, X2, · · · , X50) e B = c(Y1, Y2, · · · , Y50) contem os dados dos pacientes que tomaram a droga, antes e depois, respectivamente. Considerando que tanto A, quanto B, são advindos de uma distribuição normal, a função do R que mostra o teste para a diferença das médias populacionais dos índices de glicose em jejum, considerando confiança de 95% é
* (a) t.test(X,Y, mu=0, conf.level=0.95, alternative=”two.sided”, var.equal=TRUE).
* (b) t.test(X,Y, mu=0, conf.level=0.975, alternative=”greater”, var.equal=TRUE).
* (c) t.test(X,Y, mu=0, conf.level=0.95, alternative=”less”, paired=TRUE).
* (d) t.test(X,Y, mu=0, conf.level=0.95, alternative=”greater”, paired=TRUE).
* (e) t.test(X,Y, mu=0, conf.level=0.95, alternative=”less”, paired=FALSE).

Resposta: letra a


```
# Dados dos índices de glicose em jejum
A <- c(X1, X2, ..., X50)  # Substitua pelos dados reais do grupo que tomou a droga
B <- c(Y1, Y2, ..., Y50)  # Substitua pelos dados reais do grupo que tomou o placebo

# Realizar o teste t para duas amostras independentes
resultado_teste <- t.test(A, B, conf.level = 0.95, alternative = "two.sided", var.equal = FALSE)

# Exibir o resultado do teste
print(resultado_teste)

```
